2. What are the rules to create variables in python?

Variable names must start with a letter (a-z, A-Z) or an underscore (_).<br>
The rest of the name can include letters, digits, and underscores.<br>
Variable names are case-sensitive (var and Var are different).<br>
Python keywords cannot be used as variable names.

3. What are the standards and conventions followed for the nomenclature of variables in python to improve code readability and maintainability?

Use lowercase letters for variable names.<br>
Use underscores to separate words in a variable name (snake_case).<br>
Avoid single-character variable names (e.g., use num_count instead of n).<br>
Use meaningful and descriptive names (total_sum, user_age).<br>
Constants are typically written in UPPERCASE.<br>
Follow PEP 8 guidelines for Python code style.

4. What will happen if a keyword is used as a variable name?

Using a keyword as a variable name will result in a SyntaxError because Python treats keywords as reserved words for specific functionality.

5. For what purpose def keyword is used?

The def keyword is used to define a function in Python. It marks the beginning of the function definition, followed by the function name and parameters.

6. What is the operation of this special character \?

The backslash (\) is used as an escape character in Python, allowing special characters to be part of a string (e.g., \n for newline, \t for tab).

8. Explain the mutable and immutable data types with proper explanation & examples.

````python
# Mutable data types can be modified after creation. Example: Lists, dictionaries, sets.
my_list = [1, 2, 3]
my_list.append(4)  # Modifies the list

# Immutable data types cannot be modified after creation. Example: Tuples, strings, integers.
my_tuple = (1, 2, 3)
# my_tuple[0] = 4  # This will raise a TypeError because tuples are immutable
````

9. Write a code to create the given structure using only for loop.<br>

```
*
***
*****
*******
*********
```

````python
rows = 5
for i in range(1, rows + 1):
    print('*' * (2 * i - 1))
````

10. Write a code to create the given structure using while loop.
```
|||||||||
|||||||
|||||
|||
|
```

````python
rows = 5
while rows > 0:
    print('|' * (2 * rows - 1))
    rows -= 1
````
