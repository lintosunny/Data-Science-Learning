1. What is the role of try and exception block?

* The role of the try and except block in Python is to handle exceptions, which are errors that occur during the execution of a program.
* The try block allows you to test a block of code for errors, and the except block lets you handle the error if one occurs.


2. What is the syntax for a basic try-except block?

```python
try:
    # Code that may raise an exception
    risky_code()
except ExceptionType:
    # Code that runs if an exception occurs
    handle_exception()
```

3. What happens if an exception occurs inside a try block and there is no matching
except block?

```python
# If an exception occurs inside a try block and there is no matching except block, the program will terminate, and Python will print a traceback message indicating the error.

# Example:
try:
    print(10 / 0)  # This will raise a ZeroDivisionError
except ValueError:
    print("This won't be executed.")
```

4. What is the difference between using a bare except block and specifying a specific
exception type?

```python
# A bare except block will catch all exceptions, including system-exiting exceptions, while specifying a specific exception type will only catch that particular type of exception.

# Example of a bare except:
try:
    print(10 / 0)
except:
    print("An error occurred!")

# Example of a specific exception:
try:
    print(10 / 0)
except ZeroDivisionError:
    print("You cannot divide by zero!")
```

5. Can you have nested try-except blocks in Python? If yes, then give an example.

```python
# Yes, you can have nested try-except blocks in Python.

try:
    try:
        print(10 / 0)  # This will raise a ZeroDivisionError
    except ZeroDivisionError:
        print("Inner: You cannot divide by zero!")
except Exception:
    print("Outer: An error occurred!")
```

6. Can we use multiple exception blocks, if yes then give an example.

```python
# Yes, you can use multiple except blocks to handle different exceptions separately.

try:
    x = int(input("Enter a number: "))
    result = 10 / x
except ValueError:
    print("Invalid input! Please enter a number.")
except ZeroDivisionError:
    print("You cannot divide by zero!")
```

a. EOFError<br>

d. MemoryError<br>
e. OverflowError<br>
f. TabError<br>

a. EOFError - Raised when the input() function hits an end-of-file condition (EOF) without reading any data.<br>
b. FloatingPointError - Raised when a floating point operation fails.<br>
c. IndexError - Raised when trying to access an index that is out of range for a list or string.<br>
d. MemoryError - Raised when an operation runs out of memory.<br>
e. OverflowError - Raised when a numerical calculation exceeds the maximum limit for a numeric type.<br>
f. TabError - Raised when inconsistent use of tabs and spaces in indentation is detected.<br>
g. ValueError - Raised when a function receives an argument of the right type but inappropriate value.

8. Write code for the following given scenario and add try-exception block to it.

a. Program to divide two numbers<br>
b. Program to convert a string to an integer<br>
c. Program to access an element in a list<br>
d. Program to handle a specific exception<br>
e. Program to handle any exception

```python

# a. Program to divide two numbers
try:
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))
    result = a / b
    print(f"Result: {result}")
except ZeroDivisionError:
    print("You cannot divide by zero!")
except ValueError:
    print("Invalid input! Please enter integers.")

# b. Program to convert a string to an integer
try:
    num_str = input("Enter a number: ")
    number = int(num_str)
    print(f"Converted number: {number}")
except ValueError:
    print("Invalid input! Please enter a valid integer.")

# c. Program to access an element in a list
my_list = [1, 2, 3]
try:
    index = int(input("Enter index to access: "))
    print(my_list[index])
except IndexError:
    print("Index out of range!")
except ValueError:
    print("Invalid input! Please enter an integer.")

# d. Program to handle a specific exception
try:
    result = 10 / 0  # This will raise ZeroDivisionError
except ZeroDivisionError:
    print("Caught a specific exception: division by zero!")

# e. Program to handle any exception
try:
    print(10 / 0)  # This will raise ZeroDivisionError
except Exception as e:
    print(f"An error occurred: {e}")
```
