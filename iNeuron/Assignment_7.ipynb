2. What is the difference between the following operators:<br>
(i) ‘/’ & ‘//’<br>
(ii) ‘**’ & ‘^’

(i) / & //:

/ is the division operator, which returns the quotient as a floating-point number (e.g., 7 / 2 = 3.5).<br>
// is the floor division operator, which returns the quotient after removing the fractional part (e.g., 7 // 2 = 3).


(ii) ** & ^:

** is the exponentiation operator, which raises the number to the power of another (e.g., 2 ** 3 = 8).<br>
^ is the bitwise XOR operator, which performs a bitwise exclusive OR operation on two integers (e.g., 2 ^ 3 = 1).

3. List the logical operators.

**not:** Reverses the boolean value, i.e., True becomes False, and False becomes True.

4. Explain right shift operator and left shift operator with examples.

```python
# Right Shift Operator (>>): Shifts the bits of a number to the right by the specified number of positions. For each shift, the least significant bit is dropped.
x = 8  # Binary: 1000
result = x >> 2  # Right shift by 2 positions: 0010 (Binary) = 2
print(result)  # Output: 2


# Left Shift Operator (<<): Shifts the bits of a number to the left by the specified number of positions, filling the right with zeroes.
x = 3  # Binary: 0011
result = x << 2  # Left shift by 2 positions: 1100 (Binary) = 12
print(result)  # Output: 12
```
