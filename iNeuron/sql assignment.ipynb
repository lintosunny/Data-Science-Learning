1. which candidate process all the skills


Example input candidate table

| id | skill     |
|----|-----------|
| 1  | DS       |
| 1  | Tableau  |
| 1  | SQL      |
| 2  | R        |
| 2  | Power BI |
| 1  | Python   |

Example output

| id |
|----|
| 1  |

```sql
SELECT c.id
FROM candidate c
WHERE NOT EXISTS (
    SELECT skill
    FROM required_skills rs
    WHERE NOT EXISTS (
        SELECT 1
        FROM candidate ca
        WHERE ca.id = c.id AND ca.skill = rs.skill
    )
);


2. Query to return product info with 0 likes

product_info table example input:

| product_D  | product      |
|------------|--------------|
| 1001       | Blog        |
| 1002       | Youtube      |
| 1003       | Education    |

product_info_likes example input:

| user_id | product_id | liked_date   |
|---------|------------|--------------|
| 1       | 1001       | 19-08-2024   |
| 2       | 1002       | 18-08-2024   |

example output:

|product_id|
|----------|
| 1        |

```sql
SELECT product_id
FROM product_info
WHERE product_id IN (
  SELECT product_id
  FROM product_info_likes
)
