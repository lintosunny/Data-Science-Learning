1. What exactly is []?

[] represents an empty list, a mutable data structure in Python used to store multiple items.

2. In a list of values stored in a variable called spam, how would you assign the value 'hello' as the third value? (Assume [2, 4, 6, 8, 10] are in spam.)

You would assign 'hello' to the third value by using spam[2] = 'hello'.

3. What is the value of spam[int(int('3' * 2) / 11)]?

'3' * 2 gives '33', int('33') converts it to 33, and 33 / 11 gives 3. So, spam[3] would return 'd' (assuming spam = ['a', 'b', 'c', 'd']).

4. What is the value of spam[-1]?

The value of spam[-1] is 'd' because -1 refers to the last element in the list.

5. What is the value of spam[:2]?

The value of spam[:2] is ['a', 'b'] because it slices the list from the start up to (but not including) the index 2.

6. What is the value of bacon.index('cat')?

The value of bacon.index('cat') is 1 because 'cat' first appears at index 1 in bacon = [3.14, 'cat', 11, 'cat', True].

7. How does bacon.append(99) change the look of the list value in bacon?

After bacon.append(99), the list becomes [3.14, 'cat', 11, 'cat', True, 99].

How does bacon.remove('cat') change the look of the list in bacon?

After bacon.remove('cat'), the list becomes [3.14, 11, 'cat', True], as only the first occurrence of 'cat' is removed.

9. What are the list concatenation and list replication operators?

The list concatenation operator is + and the list replication operator is *.

10. What is the difference between the list methods append() and insert()?

append() adds an item to the end of the list, while insert() adds an item at a specific index.

11. What are the two methods for removing items from a list?

The two methods are remove() (removes the first matching value) and pop() (removes the item at a specific index or the last item by default).

12. Describe how list values and string values are identical.

Both lists and strings are sequences that allow indexing, slicing, and iteration.

13. What's the difference between tuples and lists?

Lists are mutable (their elements can be changed), whereas tuples are immutable (their elements cannot be changed).

14. How do you type a tuple value that only contains the integer 42?

You can type it as (42,). The comma is required to indicate it's a tuple.

15. How do you get a list value's tuple form? How do you get a tuple value's list form?

To convert a list to a tuple, use tuple(list_name). To convert a tuple to a list, use list(tuple_name).

16. Variables that "contain" list values are not necessarily lists themselves. Instead, what do they contain?

They contain references to the list objects in memory.

17. How do you distinguish between copy.copy() and copy.deepcopy()?

copy.copy() creates a shallow copy (only copies the outer object, not nested objects), while copy.deepcopy() creates a deep copy (copies both the outer and nested objects).
